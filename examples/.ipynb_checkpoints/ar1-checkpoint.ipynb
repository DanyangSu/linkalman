```python
import numpy as np
import pandas as pd
import linkalman
import scipy
from linkalman.models import BaseConstantModel as BCM
from linkalman.core.utils import simulated_data, ft, df_to_list, list_to_df, \
        gen_PSD, get_ergodic, create_col, clean_matrix
from linkalman.core import Filter
import nlopt
%matplotlib inline 

```

```python
def my_f(theta):
    """
    AR(1) model. In general, MLE is biased, so the focus should be 
    more on prediction fit, less on parameter estimation. The 
    formula here for Ar(1) is:
    y_t = c + Fy_{t-1} + epsilon_{t-1}
    """

    # Define theta
    phi_1 = 1 / (np.exp(theta[0])+1)
    sigma = np.exp(theta[1]) 
    sigma_R = np.exp(theta[2])

    # Generate F
    F = np.array([[phi_1]])

    # Generate Q
    Q = np.array([[sigma]])
    
    # Generate R
    R = np.array([[sigma_R]])

    # Generate H
    H = np.array([[1]])

    # Generate B
    B = np.array([[theta[3]]])

    # Collect system matrices
    M = {'F': F, 'Q': Q, 'H': H, 'R': R, 'B': B}

    return M


def my_solver(param, obj, **kwargs):
    # conform to the nlopt format
    def nlopt_obj(x, grad):
        output = obj(x)
        print(output)
        return output

    opt = nlopt.opt(nlopt.LN_BOBYQA, param.shape[0])
    opt.set_max_objective(nlopt_obj)
    opt.set_xtol_rel(1e-4)
    opt.verbose = 1
    theta_opt = opt.optimize(param)
    fval_opt = opt.last_optimum_value()
    return theta_opt, fval_opt

```

```python
# Initialize the model
model = BCM(method='LLY')
model.set_f(my_f)
model.set_solver(my_solver) 
```

```python
# Generate fake data
theta = np.array([-1, -0.1, 0.1, 1])
T = 500
cutoff_t = np.floor(T * 0.7).astype(int)
offset_t = np.floor(T * 0.9).astype(int)
x_col = ['const']
Xt = pd.DataFrame({x_col[0]: np.ones(T)})

df, y_col, xi_col = model.simulated_data(input_theta=theta, Xt=Xt, T=T)

# Create a training set
df_train = df.iloc[0:cutoff_t].copy()
df_test = df.copy()  # Keep the full data for forward prediction

# Create an offset
df_test['y_0_vis'] = df_test.y_0.copy()  # fully visible y
df_test.loc[df.index >= offset_t, ['y_0']] = np.nan
```

```python
# Fit data:
theta_init = np.random.rand(len(theta))
model.fit(df_train, theta_init, y_col=y_col, x_col=x_col)
```

```python
# Make predictions:
df_pred = model.predict(df_test)
```

```python
# Compare Kalman filter predictions against actuals
df = pd.concat([df_pred, df_test], axis=1)
df['kf_ub'] = df.y_0_filtered + 1.96 * np.sqrt(df.y_0_fvar)
df['kf_lb'] = df.y_0_filtered - 1.96 * np.sqrt(df.y_0_fvar)
df[['y_0_filtered', 'kf_ub', 'kf_lb', 'y_0_vis']].plot(figsize=(20, 10))
```

```python
df['ks_ub'] = df.y_0_smoothed + 1.96 * np.sqrt(df.y_0_svar)
df['ks_lb'] = df.y_0_smoothed - 1.96 * np.sqrt(df.y_0_svar)
df[['y_0_smoothed', 'ks_ub', 'ks_lb', 'y_0_vis']].plot(figsize=(20, 10))
```
